```python
#problem7: Mendel's First Law

k = int(input("Homo-dominant only like AA"))
m = int(input("Hetero like Aa"))
n = int(input("Homo-recessive only like aa"))

#probability that appears Homo-dominant(AA) from random mating
#전체 A의 수
#AA or Aa -> dominant phenotype

def dominant_phenotype_probability(k, m, n):
    total = k + m + n
    prob = 0
    prob += k * (k-1) #AA x AA
    prob += k * m * 2 #AA x Aa
    prob += k * n * 2#AA x aa
    prob += m * (m-1) * 0.75 #Aa x Aa
    prob += m * n * 0.5 * 2 #Aa x aa
    prob += n * (n-1) * 0 #aa x aa

    return prob/(total*(total-1))

print(dominant_phenotype_probability(k, m, n))
```

```python
#problem10: Consensus and Profile

mypath = "rosalind_cons.txt"
data = {}
list_A = list()
list_C = list()
list_G = list()
list_T = list()

with open(mypath, "r") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            sample = line
            data[sample] = ""
        else:
            data[sample] += line

    seq_length = len(next(iter(data.values())))

    for i in range(seq_length):
        a_cnt, c_cnt, g_cnt, t_cnt = 0, 0, 0, 0
        for read in data.values():
            base = read[i] 
            if "A" == base:
                a_cnt += 1
                continue
            elif "C" == base:
                c_cnt += 1
                continue
            elif "G" == base:
                g_cnt += 1
                continue
            elif "T" == base:
                t_cnt += 1
                continue
        list_A.append(a_cnt)
        list_C.append(c_cnt)
        list_G.append(g_cnt)
        list_T.append(t_cnt)

    consensus = ""
    for i in range(seq_length):
        counts = {
            "A": list_A[i],
            "C": list_C[i],
            "G": list_G[i],
            "T": list_T[i]
        }
        consensus += max(counts, key=counts.get)
    
    profile = {
    "A": list_A,
    "C": list_C,
    "G": list_G,
    "T": list_T
}
    
    print(consensus)
    for base in "ACGT":
        print(base + ": " + " ".join(map(str, profile[base])))
```

```python
#problem11: Mortal Fibonacci Rabbits
#다시 한 번 볼 것

#1개월 째 토끼: 1, 0, 0
#2개월 째 토끼: 0, 1, 0
#3개월 째 토끼: 1, 0, 1
#4개월 째 토끼: 1, 1, 0
#5개월 째 토끼: 1, 1, 1
#6개월 째 토끼: 2, 1, 1

#m개월 후에 죽음
#n번째 달
#1개월째: 새끼 1쌍
#2개월째: 성체 1쌍
#3개월째: 성체 1쌍 + 새끼 1쌍
#m개월째: 성체(m-1)쌍 + 새끼(m-2)*성체(m-2)쌍 - 성체1쌍
#n개월쨰: 성체(n-1)쌍 + 새끼(n-2)*성체(n-2)쌍 - 성체(m)쌍?

#new[1] = 1
#new[2] = 0
#new[n] = sum(new[n-2], new[n-3], ... new[n-m]) (n>=2)
#alive[n] = sum(new[n-m+1], ... new[n])

n = int(input("n<=100"))
m = int(input("m<=20"))

#첫번째 풀이

def fibonacci_rabbits(n, m):
    new = [0] * n
    new[0] = 1 #0번째 0이 1, 첫달에 한 쌍이니까

    for month in range(1, n):
        if month < m:
            new[month] = sum(new[:month-1]) #아무도 안 죽음
        else:
            new[month] = sum(new[month-m:month-1]) #죽은 세대 제외

    return sum(new[-m:]) #살아있는 토끼만


#2번째 풀이

def fibonacci_rabbits2(n, m):
    ages = [0] * m
    ages[0] = 1

    for _ in range(1, n):
        new = sum(ages[1:]) # 성체가 낳은 새끼
        ages = [new] + ages[:-1] # 가장 늙은 세대 사망
    
    return sum(ages)


print(fibonacci_rabbits(n, m))
print(fibonacci_rabbits2(n, m))
```
