```python
#problem11: Mortal Fibonacci Rabbits
#다시 한 번 볼 것

#1개월 째 토끼: 1, 0, 0
#2개월 째 토끼: 0, 1, 0
#3개월 째 토끼: 1, 0, 1
#4개월 째 토끼: 1, 1, 0
#5개월 째 토끼: 1, 1, 1
#6개월 째 토끼: 2, 1, 1

#m개월 후에 죽음
#n번째 달
#1개월째: 새끼 1쌍
#2개월째: 성체 1쌍
#3개월째: 성체 1쌍 + 새끼 1쌍
#m개월째: 성체(m-1)쌍 + 새끼(m-2)*성체(m-2)쌍 - 성체1쌍
#n개월쨰: 성체(n-1)쌍 + 새끼(n-2)*성체(n-2)쌍 - 성체(m)쌍?

#new[1] = 1
#new[2] = 0
#new[n] = sum(new[n-2], new[n-3], ... new[n-m]) (n>=2)
#alive[n] = sum(new[n-m+1], ... new[n])

n = int(input("n<=100"))
m = int(input("m<=20"))

#첫번째 풀이

def fibonacci_rabbits(n, m):
    new = [0] * n
    new[0] = 1 #0번째 0이 1, 첫달에 한 쌍이니까

    for month in range(1, n):
        if month < m:
            new[month] = sum(new[:month-1]) #아무도 안 죽음
        else:
            new[month] = sum(new[month-m:month-1]) #죽은 세대 제외

    return sum(new[-m:]) #살아있는 토끼만


#2번째 풀이

def fibonacci_rabbits2(n, m):
    ages = [0] * m
    ages[0] = 1

    for _ in range(1, n):
        new = sum(ages[1:]) # 성체가 낳은 새끼
        ages = [new] + ages[:-1] # 가장 늙은 세대 사망
    
    return sum(ages)


print(fibonacci_rabbits(n, m))
print(fibonacci_rabbits2(n, m))
```

```python
#problem14: Finding a Shared Motif
#다시 한 번 볼 것

data = {}
myfile = "rosalind_lcsm.txt"
with open(myfile) as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line[1:]
            data[id] = ""
        else:
            data[id] += line

        
    min_length = min(len(seq) for seq in data.values())
    #print(min_length) #16
    shortest = min(data.values(), key = len)
    strings = list(data.values())

    found = False
    for length in range(min_length, 0, -1): #제일 긴 것부터
        for i in range(min_length - length + 1): 
            candidate = shortest[i:i+length]
            if all(candidate in s for s in strings):
                print(candidate)
                found = True
                break
        if found:
            break
        
```

```python
#problem15: Independent Alleles
#combination 함수 다시 볼 것

#Aa Bb x Aa Bb
#k번째 세대의 N명의 Aa Bb 자식이 나올 확률
# Aa 1/2 * Bb 1/2 = 1/4
#k세대 = 2^k 자손 수
#이항분포: 확률이 p인 사건이 n번 독립적으로 발생할 때, k번 이상 성공할 확률
#확률이 1/4인 사건이 2^k번 독립적으로 발생할 때, N번 이상 성공할 확률
#N명일 때부터 2^k(max)명일 때까지
#2^k 개중에 N을 고르는 경우의 수 * p^i(1-p)^(n-i) 다 더해야됨
#1/4일 경우와 3/4일 경우를 각각 계산

#sample: 2 1 -> 0.684

k = int(input("kth generation"))
N = int(input("how many"))

#combination 함수 만들기
def combi(n, r):
    if r > n - r: # 5C3이나 5C2나 같음
        r = n - r
    result = 1
    for i in range(1, r+1):
        result *= n
        result //= i #정수 몫만 계산 # result = result // i
        n -= 1 # n x (n-1) ...
    return result


def independent_alleles(k, N):
    kth_generation = 2**k
    prob_list = list()
    
    for i in range(N, kth_generation+1):
        prob = combi(kth_generation, i) * (1/4)**i * (3/4)**(kth_generation - i) #p^i * (1-p)^(n-i)
        prob_list.append(prob)
    return sum(prob_list)

print(round(independent_alleles(k, N), 3))
```
