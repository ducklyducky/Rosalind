```python
#problem21: Locating Restriction Sites 

#fasta 파일 파싱
sequence = {}
with open("rosalind_revp.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line[1:]
            sequence[id] = ""
        else:
            sequence[id] += line

#상보적 서열 만드는 함수
def reverse_complement(seq):
    rev_seq = seq[::-1]
    rc_seq = rev_seq.replace("A", "t").replace("T", "a").replace("C", "g").replace("G", "c")
    rc_seq = rc_seq.upper()
    return(rc_seq)

def find_palindrome(sequence = dict()):
    #1based 포멧으로 몇번째 포지션인지
    #그게 몇 글자인지(4~12글자)
    positions = []
    words_length = []
    
    for read in sequence.values():
        read_length = len(read)
        for j in range(4, 13):
            for i in range(read_length - j + 1):
                window = read[i:i+j]
                if window == reverse_complement(window):
                    positions.append(i + 1) #1based
                    words_length.append(len(window))

    return positions, words_length

# find_palindrome 함수는 2가지 리스트를 인자로 반환하나
# 그것은 tuple로 묶인 상태임, zip안에는 별개의 리스트가
# 들어가야 하므로 각각 할당 다시 해줘야 함
positions, lengths = find_palindrome(sequence)
for pos, words in zip(positions, lengths):
    print(f"{pos} {words}")
```

```python
#problem22: RNA Splicing
#인트론 스플라이싱 다시 볼 것
#max(,key에 lambda 이용)

dataset = {}
with open("rosalind_splc.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line[1:]
            dataset[id] = ""
        else:
            dataset[id] += line

#remove introns
longest_id = max(dataset, key = lambda k: len(dataset[k]))
dna = dataset[longest_id]
#splicing
exon = dna
for intron_id, intron_seq in dataset.items():
    if intron_id != longest_id:
        exon = exon.replace(intron_seq, "") #rosalind10 중에서 12랑 15 인트론 서열 제거
        

#dna -> rna transcription
rna_dataset = {longest_id: exon.replace("T", "U")}

#codon table for translation
codon_table = {
    "UUU": "F", "UUC": "F",
    "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I",
    "AUG": "M",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y",
    "CAU": "H", "CAC": "H",
    "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N",
    "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C",
    "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "UAA": "*", "UAG": "*", "UGA": "*"
}

def translation(rna_dataset = dict()):
    protein = ""
    codon_started = False
    for read in rna_dataset.values():
        read_length = len(read)
        for i in range(0, read_length - 2, 3): #하나의 ORF만 보면 됨
            codon = read[i:i+3]
            aa = codon_table.get(codon, "")

            if not codon_started:
                if codon == "AUG":
                    codon_started = True
                    protein += "M"
            else:
                if aa == "*":
                    break
                elif aa != "":
                    protein += aa
    return protein

print(translation(rna_dataset))
```

```python
#problem25: Genome Assembly as Shortest Superstring
#내코드 이제 정답됨
#오류 횟수 가장 많았음.. 몇번이고 다시 좀 봐야할듯..

dna_data = {}
with open("rosalind_long.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line[1:]
            dna_data[id] = ""
        else:
            dna_data[id] += line

#2개의 read를 받아서 절반 넘게 겹치는 길이를 반환하는 함수
def overlap(read1, read2):
    max_len = min(len(read1), len(read2))
    #half_len = max_len // 2 + 1
    #작은 겹침 길이부터 시도하면 안 되고 큰 겹침 길이부터 시도해야 됨
    for l in range(max_len, 0, -1): #read2가 기준이어야 함
        suffix = read1[-l:] #read1의 끝에서 l글자
        prefix = read2[:l] #read2의 앞에서 l 글자
        if suffix == prefix:
        #아니면 이렇게 한 줄로도 가능
        #if read1.endswith(read2[:l]):
        #혹은
        #if read1[-l:] == read2[:l]
            return l #겹침 길이 반환
    
    return 0 #겹침이 없는 경우

#overlap을 사용한 assemble 함수
def assemble_strings(dna_data: dict):
    reads = list(dna_data.values())
    
    #겹치는 read를 계속 병합, read가 1개 남을 때까지 반복
    #병합은 2개 이상이어야 하기 때문
    while len(reads) > 1:
        max_overlen = -1 #발견된 최대 겹침 길이
        best_pair = (None, None) #가장 많이 겹치는 두 read의 index
        merged_read = ""     #병합된 read 저장

        #모든 read 쌍을 비교
        for i in range(len(reads)):
            for j in range(len(reads)):
                if i == j:
                    continue
                a, b = reads[i], reads[j]
                overlen = overlap(a, b) #overlap으로 겹침길이 계산
                if overlen > max_overlen: #가장 많이 겹치는 걸 저장
                    max_overlen = overlen
                    best_pair = (i, j)
                    merged_read = a + b[overlen:] #겹치는 부분만 이어붙임
        if max_overlen <= 0:
            break

        #병합된 이후로 read 업데이트
        #병합에 쓰인 두 read 제거
        i, j = best_pair
        new_reads = []
        for k in range(len(reads)):
            if k != i and k != j:
                new_reads.append(reads[k]) #병합에 안쓰인 것들만 read 추가
        new_reads.append(merged_read) #현재 병합된 read 추가
        #아니면 list comprehension
        #new_reads = [reads[k] for k in range(len(reads)) if k not in (i, j)]
        reads = new_reads #업데이트

    return reads[0]

print(assemble_strings(dna_data).strip())
```

```python
#problem25: Genome Assembly as Shortest Superstring
#모니카 코드

dna_data = {}
with open("rosalind_long.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line[1:]
            dna_data[id] = ""
        else:
            dna_data[id] += line


def overlap(read1, read2):
    max_len = min(len(read1), len(read2))
    for l in range(max_len, 0, -1):  # 최대 길이부터 시작
        if read1[-l:] == read2[:l]:  # read1의 끝과 read2의 시작 비교
            return l  # 겹침 길이 반환
    return 0  # 겹침이 없는 경우

def assemble_strings(dna_data: dict):
    """ DNA 조각들을 병합하여 하나의 문자열로 만드는 함수 """
    reads = list(dna_data.values())
    
    while len(reads) > 1:
        max_overlen = -1
        best_pair = (None, None)
        merged_read = ""

        for i in range(len(reads)):
            for j in range(len(reads)):
                if i != j:  # 같은 인덱스는 비교하지 않음
                    overlen = overlap(reads[i], reads[j])
                    if overlen > max_overlen:
                        max_overlen = overlen
                        best_pair = (i, j)
                        merged_read = reads[i] + reads[j][overlen:]  # 겹침 부분 제외하고 병합

        if max_overlen <= 0:  # 더 이상 겹치는 부분이 없으면 종료
            break

        # 병합된 이후로 read 업데이트
        i, j = best_pair
        reads = [reads[k] for k in range(len(reads)) if k != i and k != j]  # 병합에 사용되지 않은 read들
        reads.append(merged_read)  # 병합된 read 추가

    return reads[0]

# 결과 출력
result = assemble_strings(dna_data).strip()
print(result)

```
