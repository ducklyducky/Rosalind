```python
#problem36: k-Mer Composition
#다시 한 번 볼 것

import itertools

read = ""
with open("rosalind_kmer.txt") as handle:
    for line in handle:
        line = line.strip()
        if not line.startswith(">"):
            read += line

base = ["A", "C", "G", "T"] #처음부터 사전순

kmers = ["".join(x) for x in itertools.product(base, repeat=4)]

kmer_cnt = {kmer: 0 for kmer in kmers}
for i in range(0, len(read)-3):
    window = read[i:i+4]
    if window in kmer_cnt:
        kmer_cnt[window] += 1

for kmer in kmers:
    print(kmer_cnt[kmer], end = " ")
```

```python
#problem37: Speeding Up Motif Finding
#다시 볼 것

read37 = ""
with open("rosalind_kmp.txt") as handle:
    for line in handle:
        line = line.strip()
        if not line.startswith(">"):
            read37 += line

def failure_array(read37):
    length = len(read37)
    Pvalue = [0] * length
    for i in range(1, length): #0번째는 항상 0
        j = Pvalue[i - 1] #이전 위치에서 얼마나 일치했는가부터 시작
        while j > 0 and read37[i] != read37[j]: #일치가 깨짐
            j = Pvalue[j - 1] #이전의 접두사 내부에서 최대 일치 길이
        if read37[i] == read37[j]: #현재 문자가 일치하면
            j += 1 #일치 길이를 늘림
        Pvalue[i] = j #현재까지 일치한 prefix suffix 길이
    return Pvalue

Pvalue = failure_array(read37)
print(" ".join(map(str, Pvalue)))
```

```python
#problem38: Finding a Shared Spliced Motif
#꼭 다시 보기
#DP(dynamic programming)동적 프로그래밍 사용
#복잡한 문제를 작은 문제로 나누고, 그 해답을 재활용해서 전체 문제를 푸는 방법
#쓰는 조건: 최적부분구조 - 문제를 더 작은 하위 문제로 나눌 수 있고 그걸 이용해서 전체 문제 답 구할 수 있음
##중복되는 문제 - 똑같은 계산이 여러 번 반복
#dp 테이블에 작은 문제의 답들을 저장함
#그냥 냅다 비교하면 경우의 수가 지수적이라서 너무 오래 걸림
#행에는 s, 열에는 t를 넣은 표를 만들고 지금까지 맞은 수로 채움
#dp[i][j]는 s[0:i]와 t[0:j]의 최장 공통 부분 수열의 길이

#표에서 맞은 글자를 거꾸로 따라가서 문자열을 만듦
#동적

data = {}
with open("rosalind_lcsq.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith(">"):
            id = line
            data[id] = ""
        else:
            data[id] += line

sample = [key for key in data.keys()]
s = data[sample[0]]
t = data[sample[1]]

#dp 만들기
def lcs(s, t):
    m, n = len(s), len(t)
    dp = [[0]*(n+1) for _ in range(m+1)]

    #안에 채우기 #LCS의 길이정보만 구할 수 있음
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]: #일치하면
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    #LCS 복원(역추적) #이걸 해야 실제 문자열을 구할 수 있음
    i, j = m, n
    lcs_str = []
    while i > 0 and j > 0:
        if s[i-1] == t[j-1]:
            lcs_str.append(s[i-1])
            i -= 1
            j -= 1

        elif dp[i-1][j] >= dp[i][j-1]:
            i -= 1
        else:
            j -= 1

    return "".join(reversed(lcs_str))

print(lcs(s, t))
```
